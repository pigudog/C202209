```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("table1_2.csv",encoding="gb2312")
df
```

```python
## labelencoder
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
df1 = df[["纹饰","类型","表面风化"]]
df1
```

```python
df1["纹饰"] = LabelEncoder().fit_transform(np.array(df1["纹饰"])).reshape(-1,1)
df1["表面风化"] = LabelEncoder().fit_transform(np.array(df1["表面风化"])).reshape(-1,1)
df1["类型"] = LabelEncoder().fit_transform(np.array(df1["类型"])).reshape(-1,1)
df1
```

```python
# 计算相关性矩阵
corr_matrix=df1.corr()
corr_matrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

sns.heatmap(corr_matrix, annot=True)
```

```python
df2 = df[["纹饰","类型","颜色","表面风化"]]
df2.head()
```
